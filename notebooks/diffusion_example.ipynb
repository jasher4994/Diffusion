# Diffusion and Stable diffusion

Diffusion models can be thought of a two-step process. 

The forward process: The gradual adding of noise to images in timesteps such that a model can distinguish two images by the noise added in a timestep. 

The reverse process: A trained neural network learns to reverse the process such that it can take pure noise and incrementally reduce noise to the point where an actual image appears. 

<b> So, effectively, we are adding noise to an image. And then using the images with different noise levels at different timestamps we are training a model to be able to reverse engineer that process. 

```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompt = "a photo of an astronaut riding a horse in scotland"
image = pipe(prompt).images[0]  
display(image)
```
